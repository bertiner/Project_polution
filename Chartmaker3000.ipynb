# Yearly

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator, MultipleLocator
import pandas as pd
```

```python
df = pd.read_excel('year_per_zone-copy1.xlsx')
#df = pd.read_excel('summer_per_zone.xlsx')
#df = pd.read_excel('autumn_per_zone.xlsx')
#df = pd.read_excel('winter_per_zone.xlsx')
#df = pd.read_excel('spring_per_zone.xlsx')

print(df.head(1))
```

## Input data

```python
orientation = df[["Orientation"]];

orientation_ticks = np.linspace(0, 390,num=13,endpoint=False);
#print(np.linspace(0, 390,num=13,endpoint=False))

plt.rcParams['lines.linewidth'] = 1.5 # Curve Thickness
matplotlib.rcParams.update({'font.size': 15}) # Font size

total_energy = df.iloc[:,-1:];
col_names = df.columns;

print(col_names.values)

total_idxs = [20]
all_idxs = list(range(1, 20));

graf1_idxs = [4,5,8,10,12,13];
graf2_idxs = [14,15,16,17,18,19];
graf3_idxs = [1,2,3];
graf4_idxs = [6,7,9];

print("Total:", col_names[total_idxs].values)
print("All:", col_names[all_idxs].values)
print("Graf 1:", col_names[graf1_idxs].values)
print("Graf 2:", col_names[graf2_idxs].values)
print("Graf 3:", col_names[graf3_idxs].values)
print("Graf 4:", col_names[graf4_idxs].values)

colors = ["#ffffff", # offset 0          
          "#EA2700", # A      1    
          "#FAEA5A", # B      2    
          "#4C6CA9", # C      3    
          "#EA2700", # D      4    
          "#EF920E", # E      5    
          "#EA2700", # F      6    
          "#FAEA5A", # G      7    
          "#FAEA5A", # H      8    
          "#4C6CA9", # I      9    
          "#A0C979", # J      10    
          "#4A708B", # K1     11    
          "#9FBDEF", # K2     12     
          "#4C6CA9", # K3     13     
          "#EA2700", # L1     14     
          "#EF920E", # L2     15     
          "#FAEA5A", # L3     16     
          "#A0C979", # M      17    
          "#9FBDEF", # N1     18     
          "#4C6CA9"]; # N2    19  
```

### Total One Curve

```python
ax = plt.figure(figsize=(9,4.2)).gca()

ax.plot(orientation.values,total_energy.values )
ax.set_xlabel("Orientation")
#ax.set_ylabel("Energy [kWh/year]")
#ax.set_ylabel("Energy [kWh/summer]")
#ax.set_ylabel("Energy [kWh/autumn]")
#ax.set_ylabel("Energy [kWh/winter]")
#ax.set_ylabel("Energy [kWh/spring]")
ax.set_xticks(orientation_ticks)
plt.xticks(rotation=0)

plt.title("Total Radiation for all surfaces - Yearly")
plt.grid(visible=True, which='major', color='grey', linestyle='-')
plt.grid(visible=True, which='minor', color='lightgrey', linestyle='-', alpha=0.5)

plt.minorticks_on()

#plt.savefig("Total.jpg", format='jpg', dpi=300, bbox_inches='tight')
plt.show()
```

### Total All Curves

```python
fig_height = 8;
fig_width = 16;
ax = plt.figure(figsize=(fig_width,fig_height),).gca()
for i in all_idxs:
    ax.plot(orientation.values,df.iloc[:,i],label=col_names[i],color=colors[i])
    ax.set_xticks(orientation_ticks)
    ax.set_xlabel("Orientation")
    #ax.set_ylabel("Energy [kWh/year]")
    #ax.set_ylabel("Energy [kWh/summer]")
    #ax.set_ylabel("Energy [kWh/autumn]")
    #ax.set_ylabel("Energy [kWh/winter]")
    #ax.set_ylabel("Energy [kWh/spring]")
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))


plt.title("Total")

# Major & Minor grid
plt.grid(visible=True, which='major', color='grey', linestyle='-')
plt.grid(visible=True, which='minor', color='lightgrey', linestyle='-', alpha=0.5)

plt.minorticks_on()
# plt.savefig("Total.jpg", format='jpg', dpi=300, bbox_inches='tight') # for å lagre figur
plt.show()

```
